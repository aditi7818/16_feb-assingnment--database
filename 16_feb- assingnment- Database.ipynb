```python
Question: Explain Primary Key and Foreign Key.
Answer:
Primary key:
A Primary key is used to ensure that data in the specific column is unique. A Column cannot have NULL value. it is either an existing table column or a column that is specifically 
generated by the database acoording to a deined sequence. or In the other the words, The primary key is a unique or non- null key that uniquely identifies every record in a table
or relation. Each database needs a unique identifier for every row of a table, and the primary keys play a vital role in identifying rows in the table uniquely. the primary key 
column cannot store duplicate values. it is also called a minimal super key; therefore, we cannot specify more than one primary key in any relationship.

For example, We have a table named customer with attributes such as ID, Name, and City. Only the ID column can never can never contain duplicate and NULL values because each customer 
has a unique identification number. This feature helps to identify each record in the database uniquely. hence, we can make ID attribute a primary key.


Foregin key: 
The foreign key is a group of one or more columns in  a database to uniquely identify another database record in some other table to maintain the referential integrity. it is also 
known as the referencing key that establishes a relationship between two different tables in a database. A foreign key always matches the primary key column in another table.
it means a foreign key column in one table refers to the primary key column of another table. it means a foregin key column in one table refers to the primary key column of another 
table. A foreign key is benefical in relational database normalization, especially when we need to access records from other tables.
A foreign key creates a parent-child relationship with the tables where the parent table holds the initial column values, and the child table refernces the parent column values.
We can achieve this relationship only when the foreign key constraint is found on the child table.

Example of Foreign Key
Let's disucss an example to understand the working of a foreign key.
Consider two tables Student and Department having thier respective attributes as shown in the below table structure:
    
                                                  Student
        
Stud_Id                                        Name                             Course

101                                            John                             Computer
105                                            Merry                            AI
107                                            Sheero                           Biology
108                                            Bisle                            Maths




                                                Department

Dept_name                                                              Stud_Id

CS_Department                                                          105
CS_Department                                                          101
Science_Department                                                     101
Math_Department                                                        108


In the tables, one attribute, you can see, is common, that is Stud_Id, but it has different key constraints for both tables. IN the Student table, the field Stud_Id is a primary key
because it is uniquely identifying all other fields of the student table. On the other hand, Stud_Id is a foreign key attribute for the department table are linked with one another 
because of the Stud_Id attribute.
```
